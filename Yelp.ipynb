# Analysis

```python
bus_rev.head()
```

```python
bus_rev.groupby('Cat_Summary')['stars_y'].mean()
```

```python
bus_rev['Cat_Summary'].value_counts()
```

```python
star = bus_rev.groupby(['Cat_Summary', 'business_id', 'name', 'review_count'], as_index=False)['stars_y'].mean()
```

```python
star
```

```python
star.groupby(['Cat_Summary'])['stars_y'].nlargest(3).reset_index()
```

```python
pd.crosstab(star['stars_y'], star['Cat_Summary'])
```

```python
star.rename(columns = {'stars_y':'ave_star'}, inplace = True)
```

```python
star.head()
```

```python
bus_rev.head()
```

```python
bus_rev.drop(columns=['useful_count'], inplace = True)
```

```python
bus_rev['pos_review'] = bus_rev['stars_y'].apply(lambda x: 1 if x > 3 else 0)
```

```python
a = bus_rev['useful']
bus_rev['pos_useful'] = np.where(bus_rev['pos_review'] == 1, a, 0)
```

```python
bus_rev
```

```python
bus_rev = pd.merge(bus_rev, user[['user_id', 'fans']], how = 'left', on = 'user_id')
```

```python
bus_rev
```

```python
count = bus_rev.groupby(['business_id']).sum()
```

```python
count.head()
```

```python
bus_star = pd.merge(star,count[['pos_review', 'pos_useful', 'useful', 'fans']],on='business_id', how='left')
```

```python
bus_star
```

```python
bus_star.groupby(['Cat_Summary'])['review_count'].quantile(0.75).reset_index()
```

```python
m = (
    bus_star.groupby(['Cat_Summary'])['review_count']
    .transform(lambda s: s.between(s.quantile(0.75), s.quantile(1.0)))
)
df = bus_star[m]
```

```python
df.groupby(['Cat_Summary'])['review_count'].min()
```

```python
df["rank"] = df.groupby("Cat_Summary")["ave_star"].rank("dense", ascending=True)
```

```python
df
```

```python
all_cat = df['Cat_Summary'].unique()
result_group = pd.DataFrame()

for i in all_cat:
    all_group = df.groupby('Cat_Summary').get_group(i)
    rank_group = all_group.nlargest(20, 'rank')
    useful_group = rank_group.nlargest(5, 'pos_useful')
    fans_group = useful_group.nlargest(5, 'fans')
    result_group = pd.concat([result_group,fans_group])

result_group
```

```python
result_group.to_csv('result.csv', index=False)
```
